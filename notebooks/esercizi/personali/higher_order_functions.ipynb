# Higher Order Functions
---
## fold_left vs fold_right

````ocaml
List.fold_right (fun x acc -> x :: acc) [1;2;3] [];;
````

````ocaml
List.fold_left (fun acc x -> x :: acc) [] [1;2;3];;
````

- fold_left, si comporta in modo simile a una reduce, l'accumulatore viene subito passato come parametro alla funzione e aggiornato di volta in volta

(nota che 1, 2, 3 sono gli elementi della lista sul quale è applicata ```fold_left```)

```acc -> acc1 = f acc 1 -> acc2 = f acc1 2 -> acc3 = f acc2 3```

acc3 è il risultato


- fold_right, la prima funzione ad essere eseguita è quella più interna

```f 1 ( f 2 ( f 3 acc0 ))```
